## solution to problem 1

```python
# lets create a function called prime_num and takes in just an input n which will return the nth term of sequence
# of prime numbers


# 1. create a list called 'value', all that gets appended to this list are prime numbers
# 2. intialize an integer that keeps incrementing which will always undergo a test to check if it prime or not.
#    if it is prime then it gets appended to value which is a list.

# there is going to be a while loop that keeps looping until the size of the list of value is equal to the nth 
# term we requested for.

def prime_num(n):
    value = [2] # list of prime numbers
    count = 2 # initialized integer that keeps incrementing
    while len(value) < n:
        test = [] # for every prime number that is a factor of count a boolean gets appended here.
        for k in value:
            if count % k == 0: # test if each number is divisible by a prime number present in value
                test.append(True) # for prime k is a factor of count, then let True be appended to test.
            else:
                test.append(False) # for prime k which is not a factor of count, then let False be appended to test.
        if True in test: # if any prime number can divide count, then count is not a prime number just increment
            count += 1
        else:
            value.append(count) # since count is not divisible by k, then it is a prime number
            count += 1
    return value[-1]

#The 50th prime number is
prime_num(50)
```

```python
# lets create a function called fib_n, that is the fib number of nth term
def fib_n(n): # find the nth value of the fibonacci sequence
    value = [0,1] # initialize the fibonacci sequence with 0 and 1
    for k in range(n-2): # iterate until we get the nth term
        new_val = value[-2] + value[-1] # add the last two term of the fib sequence
        value.append(new_val) # append the new value to the fib sequence (update)
    return value[-1] # return the nth value

fib_n(100)
```
